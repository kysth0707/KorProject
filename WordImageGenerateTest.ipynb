단어 입력하면 이미지 생성해주는 예제

```python
from PIL import Image, ImageFont, ImageDraw
```

```python
fontPath = r'E:\GithubProjects\KorProject\KorProject\fonts\MYYeongnamnu.ttf'
```

```python
def generateImage(text : str, fontSize : int = 60, IMAGE_SIZE : tuple = (1920,1080)):
	IMAGE_WIDTH, IMAGE_HEIGHT = IMAGE_SIZE

	font = ImageFont.truetype(fontPath, fontSize)

	img = Image.new(mode="L", size=(IMAGE_WIDTH, IMAGE_HEIGHT), color=255)
	drawing = ImageDraw.Draw(img)
	_, _, w, h = drawing.textbbox((0, 0), "응아니야", font=font)
	drawing.text(
		((IMAGE_WIDTH-w)/2, (IMAGE_HEIGHT-h)/2),
		"응 아니야",
		fill=0,
		font=font,
	)
	return img
```

```python
generateImage("응 아니야", 450)
```

```python
savePath = r'E:\GithubProjects\KorProject\KorProject\imgs\yesno.png'
generateImage("응 아니야", 450).save(savePath)
```
