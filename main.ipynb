```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
```

```python
fontPath = r'E:\GithubProjects\KorProject\KorProject\fonts\HANBatang.ttf'
```

```python
data = []
with open('dataset.txt', encoding='utf8') as f:
	while True:
		line = f.readline()
		if not line:
			break
		a,b,c = line.rstrip().split(' ')
		data.append((
			a.split(','),
			int(b),
			float(c[:-1]),
		))
data
```

https://stackoverflow.com/questions/38465478/wordcloud-from-data-frame-with-frequency-python

```python
showData = {}
for item in data:
	texts, count, percent = item
	showData[texts[0]] = count
```

```python
wc = WordCloud(font_path=fontPath,
               background_color="white",
               max_font_size=60)
```

```python
cloud = wc.generate_from_frequencies(frequencies=showData)

plt.figure(figsize=(16, 9))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
import seaborn as sns
```

```python
def showGraph(start,end,rotationValue : int = 0):
	colors = sns.color_palette('gist_rainbow', end - start)
	plt.figure(figsize=(12, 8))
	plt.ylabel('빈도수 합')
	plt.xticks(rotation=rotationValue)
	plt.xlabel(f'비속어 [ 범위 : {start} ~ {end - 1} ]')
	plt.bar(list(showData.keys())[start:end], 
			list(showData.values())[start:end],
			color = colors,
	)
```

```python
showGraph(1, 68, 90)
```

```python
showGraph(1, 20, 30)
```
