```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import random
```

```python
fontPath = r'E:\GithubProjects\KorProject\KorProject\fonts\HANBatang.ttf'
```

```python
# texts = list("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789")


# data = [random.choice(texts) for _ in range(7000)]
# data

textPath = r'E:\GithubProjects\KorProject\KorProject\texts\unigram_freq.txt'

data = []
with open(textPath, 'r') as f:
	while True:
		line = f.readline()
		if not line:
			break
		data.append(line.split(',')[0])
data = data[:400]
```

```python
wc = WordCloud(font_path=fontPath,
               background_color="white",
               max_font_size=60)
```

```python
cloud = wc.generate(' '.join(data))

plt.figure(figsize=(10, 8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
imagePath = r'E:\GithubProjects\KorProject\KorProject\imgs\yesno.png'
```

```python
imageMask = np.array(Image.open(imagePath))
imageMask
```

```python
imageMask.shape
```

```python
wc = WordCloud(font_path=fontPath,
               background_color="white",
               max_font_size=60,
               mask=imageMask,)
```

```python
cloud = wc.generate(' '.join(data))

plt.figure(figsize=(10, 8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```

```python
wc = WordCloud(font_path=fontPath,
               background_color="white",
               max_font_size=60,
               mask=imageMask,
               contour_width=3,
               contour_color="firebrick")
```

```python
cloud = wc.generate(' '.join(data))

plt.figure(figsize=(10, 8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
